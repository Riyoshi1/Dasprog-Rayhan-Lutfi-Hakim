```python
# 3 IF ELIF ELSE
S1, S2, S3 = map(str, input().split())
S1 = len(S1)
S2 = len(S2)
S3 = len(S3)

if S1 == S2 or S1 == S3 or S2 == S3:
    print("Hebwueh")
elif S1 < S2 > S3 or S1 > S2 < S3:
    print("Yeehaaw")
else:
    print("Woohoohoo")
```

```python
# 5 HURUF DIKALI
N = input()
total = 1
for i in N:
    if i.isalpha():
        i = i.lower()
        total *= (ord(i) - 96)
    elif i.isdigit():
        total *= int(i)
print(total)
```

```python
# 7 Panji ingin mencetak sebuah string sepanjang N dengan pola sebagai berikut #1
N = int(input())
count = 0
pola = ["##**", "**##"]
while count < N:
    for i in range(2):
        pola_akhir = (pola[(i % 2) - 1] * ((N // len(pola[i % 2])) + 1))[:N]
        count += 1
print(pola_akhir, end="")
```

```python
# 8 Panji ingin mencetak format matriks berukuran N x M dengan pola #1
N, M = map(int, input().split())
pola_genap = '**##'
pola_ganjil = '##**'

for i in range(N):
    pola = pola_genap if i % 2 == 0 else pola_ganjil
    simbol = (pola * ((M // len(pola)) + 1))[:M]
    print(simbol)
```

```python
# 9 Panji ingin mencetak sebuah string sepanjang N dengan pola sebagai berikut #2
N = int(input())
if N % 2 == 0:
    for i in range(N):
        if i % 2 == 0:
            print("*", end="")
        else:
            print("#", end="")
elif N % 2 != 0:
    for j in range(N):
        if j % 2 == 0:
            print("#", end="")
        else:
            print("*", end="")
```

```python
# 10 Panji ingin mencetak format matriks berukuran N x M dengan pola #2
N, M = map(int, input().split())
for i in range(N):
    for j in range(M):
        if (i + j) % 2 == 0:
            print("*", end="")
        else:
            print("#", end="")
    print()
```

```python
# 12 Info Geser
def geser(huruf, N):
    if huruf.isalpha():
        if huruf.islower():
            huruf_baru = (ord(huruf) - ord('a') + N) % 26 + ord('a')
            return chr(huruf_baru)
        elif huruf.isupper():
            huruf_baru = (ord(huruf) - ord('A') + N) % 26 + ord('A')
            return chr(huruf_baru)
    else:
        return huruf

enkripsi = input()
validasi = input()

ketemu = False
for N in range(1, 26):
    hasil = ''
    for huruf in enkripsi:
        hasil += geser(huruf, N)
    hasil = hasil[::-1]
    if hasil == validasi:
        print(f"Silahkan berjalan sejauh {N} meter")
        ketemu = True
        break

if not ketemu:
    print("Mending pulang aja")
```
