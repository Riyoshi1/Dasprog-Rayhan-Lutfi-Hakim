```python
# Katamari
N = int(input())
i = 0
while True:
    if N == 0:
      print("Maybe it's better to add 1 fences, just to be perfect.")
      break
    elif 2**i - 1 == N:
        print("It's the perfect amount, nice!")
        break
    elif 2**i - 1 > N:
        if (2**i - 1) - N < N - (2**(i-1) - 1):
            print(f"Maybe it's better to add { (2**i - 1) - N } fences, just to be perfect.")
            break
        else:
            print(f"Maybe it's better to remove { N - (2**(i-1) - 1) } fences, just to be perfect.")
            break
    i += 1
```

```python
# Circle
import math
x1, y1, r1 = map(float, input().split())
x2, y2, r2 = map(float, input().split())
d = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)  # menghitung jarak antar 2 titik pusat
if d < r1 + r2 and d > abs(r1 - r2):
    a = (r1**2 - r2**2 + d**2) / (2 * d) # Jarak dari pusat lingkaran pertama ke titik di antara titik potong
    h = math.sqrt(r1**2 - a**2) # Jarak dari titik tengah x0, y0 ke masing-masing titik potong
    x0 = x1 + a * (x2 - x1) / d
    y0 = y1 + a * (y2 - y1) / d
    xa = x0 + h * (y2 - y1) / d
    ya = y0 - h * (x2 - x1) / d
    xb = x0 - h * (y2 - y1) / d
    yb = y0 + h * (x2 - x1) / d
    print("Tidaaaaaak, airnya muncrat!")
    print(f"{xa + ya + xb + yb:.2f}")
elif d == r1 + r2 or d == abs(r1 - r2):
    a = (r1**2 - r2**2 + d**2) / (2 * d)
    x0 = x1 + a * (x2 - x1) / d
    y0 = y1 + a * (y2 - y1) / d
    print("Fyuh, hampir saja")
    print(f"{x0:.2f} {y0:.2f}")
else:
    print("Amang ges")
```

```python
N = int(input())
count = 0
pola = ["##**", "**##"]
while count < N:
    for i in range(2):
        pola_akhir = (pola[(i % 2) - 1] * ((N // len(pola[i % 2])) + 1))[:N]
        count += 1
print(pola_akhir, end="")
```

```python
# 5
n = int(input()) # banyak segitiga
hasil = []
for i in range(n):
  baris = int(input())
  segitiga = []
  for j in range(baris):
    isi = list(map(int, input().split()))
    segitiga.append(isi)
  for a in range(baris - 2, -1, -1):
    for b in range(len(segitiga[a])):
      segitiga[a][b] += max(segitiga[a + 1][b], segitiga[a + 1][b + 1])
  hasil.append(segitiga[0][0])

print(' '.join(map(str, hasil)))
```
